# Sharpening

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
img = cv2.imread('./images/einstein2.jpg', cv2.IMREAD_GRAYSCALE)  # You can use any image path here

# Define your custom kernel (e.g., a sharpening filter)
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Apply convolution using filter2D
convolved_img = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)

# Plotting the original and convolved images using Matplotlib
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis ticks

# Convolved Image
plt.subplot(1, 2, 2)
plt.imshow(convolved_img, cmap='gray')
plt.title('Convolved Image (Sharpening)')
plt.axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

# Smoothing (Gaussian Blur)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
img = cv2.imread('./images/einstein2.jpg', cv2.IMREAD_GRAYSCALE)  # You can use any image path here


# Apply convolution using filter2D
convolved_img = cv2.GaussianBlur(img, (15, 15), 0)  # Kernel size (5, 5) and sigma=0

# Plotting the original and convolved images using Matplotlib
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis ticks

# Convolved Image
plt.subplot(1, 2, 2)
plt.imshow(convolved_img, cmap='gray')
plt.title('Convolved Image (Blur)')
plt.axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

# Edge Detection (Sobel Filter)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
img = cv2.imread('./images/einstein2.jpg', cv2.IMREAD_GRAYSCALE)  # You can use any image path here

# Define your custom kernel 
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# Apply convolution using filter2D
convolved_img = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)

# Plotting the original and convolved images using Matplotlib
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis ticks

# Convolved Image
plt.subplot(1, 2, 2)
plt.imshow(convolved_img, cmap='gray')
plt.title('Convolved Image (Sobel)')
plt.axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

# Edge Detection (Canny)

```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Load the image
img = cv2.imread('./images/einstein2.jpg', cv2.IMREAD_GRAYSCALE)  # You can use any image path here

# Step 2: Convert the image to grayscale
# gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 3: Apply Gaussian Blur to reduce noise
# blurred_img = cv2.GaussianBlur(img, (5, 5), 1.5)

# Step 4: Apply Canny edge detection
edges = cv2.Canny(img, threshold1=100, threshold2=200)

# Step 5: Plotting using Matplotlib
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct color display
plt.title('Original Image')
plt.axis('off')  # Hide axis ticks

# Canny Edges
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.imshow(edges, cmap='gray')  # Display edges in grayscale
plt.title('Canny Edges')
plt.axis('off')  # Hide axis ticks

# Display the plot
plt.tight_layout()
plt.show()
```

# Edge Detection (Prewitt)

```python
Prewitt Operator
The Prewitt operator uses two convolution kernels to approximate the gradient of the image intensity. The two kernels are:

One for detecting horizontal edges.
One for detecting vertical edges.

```

<img src="./images/Prewitt.png" alt="Prewitt" width="1000"/>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image
img = cv2.imread('./images/einstein2.jpg', cv2.IMREAD_GRAYSCALE)  # You can use any image path here

# Step 2: Define Prewitt kernels
prewitt_kernel_x = np.array([[-1, 0, 1],
                              [-1, 0, 1],
                              [-1, 0, 1]])

prewitt_kernel_y = np.array([[1, 1, 1],
                              [0, 0, 0],
                              [-1, -1, -1]])

# Step 3: Apply convolution using the Prewitt kernels
edges_x = cv2.filter2D(img, -1, prewitt_kernel_x)  # Horizontal edges
edges_y = cv2.filter2D(img, -1, prewitt_kernel_y)  # Vertical edges

# Step 4: Combine the edges
edges = np.sqrt(edges_x**2 + edges_y**2)
edges = np.uint8(edges)  # Convert back to uint8 type

# Step 5: Plotting the original and edge-detected images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis ticks

# Prewitt Edges
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Prewitt Edges')
plt.axis('off')  # Hide axis ticks

# Display the plot
plt.tight_layout()
plt.show()
```
