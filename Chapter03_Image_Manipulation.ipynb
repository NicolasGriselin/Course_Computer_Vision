## Loading and Showing

```python
import cv2
import matplotlib.pyplot as plt

# 1. Load the image using OpenCV
image = cv2.imread('./images/obamas.jpg')  # Replace with your image path

# 2. Get image properties
height, width, channels = image.shape  # Shape gives (height, width, channels)
dtype = image.dtype  # Data type of the image (e.g., uint8)
total_pixels = image.size  # Total number of pixels (height * width * channels)

# 3. Print image properties
print(f"Image Dimensions: {width} x {height} pixels")
print(f"Number of Channels: {channels}")  # For grayscale, this will not exist
print(f"Data Type: {dtype}")  # E.g., uint8 (8-bit)

# 4. Plot the image using Matplotlib
plt.imshow(image)
plt.title('Loaded Image')
plt.axis('off')  # Turn off axis labels
plt.show()

```

```python
# cv2.imread(filename, flags)

# filename: The path to the image file (string).
# flags: Specifies how the image should be read. This is an optional parameter, and its default value is cv2.IMREAD_COLOR (1).
```

```python
# cv2.IMREAD_COLOR (1) - Load as a color image (default):

# This is the default behavior if no flag is provided.
# The image is loaded as a color image with three channels (BGR format).
# Any transparency channel present in the image is discarded.

import cv2

# Load image as color
image_color = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)

# OpenCV loads images in BGR format by default, we need to convert it to RGB for correct display in matplotlib
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Plot the image using Matplotlib
plt.imshow(image_rgb)
plt.title('Loaded Image')
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
# 2. cv2.IMREAD_GRAYSCALE (0) - Load as a grayscale image:

# Loads the image in grayscale (single channel, 8-bit depth).
# This option is useful for tasks like edge detection or when the color information is not needed.

# Load image as grayscale
image_gray = cv2.imread('./images/obamas.jpg', cv2.IMREAD_GRAYSCALE)

# Plot the image using Matplotlib
plt.imshow(image_gray)
plt.title('Loaded Image')
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
# To select and apply a colormap when displaying an image in Matplotlib, you can pass the cmap parameter to plt.imshow().

# Available Colormaps:

# Grayscale: gray or Greys
# Heatmap: hot, inferno, plasma, magma
# Rainbow-like: jet, viridis, turbo
# Binary: binary
# Cool-Warm: cool, coolwarm
# You can find the full list of available colormaps in Matplotlib here.

# Grayscale:
plt.imshow(image_gray, cmap='gray')  # Grayscale colormap
```

```python
# Heatmap Colormap:
plt.imshow(image_gray, cmap='hot')  # Heatmap style
```

## Resize

```python
# Load image as color
image_color = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)

# OpenCV loads images in BGR format by default, we need to convert it to RGB for correct display in matplotlib
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Get image properties
height, width, channels = image_rgb.shape  # Shape gives (height, width, channels)
dtype = image_rgb.dtype  # Data type of the image (e.g., uint8)
total_pixels = image_rgb.size  # Total number of pixels (height * width * channels)

# Print image properties
print(f"Image Dimensions: {width} x {height} pixels")
print(f"Number of Channels: {channels}")  # For grayscale, this will not exist
print(f"Data Type: {dtype}")  # E.g., uint8 (8-bit)

# Plot the image using Matplotlib
plt.imshow(image_rgb)
plt.title('Loaded Image')
plt.axis('off')  # Turn off axis labels
plt.show()


# Resize the image to 80x60 pixels
resized_image = cv2.resize(image_rgb, (600, 600))  # (width, height)

# Plot the image using Matplotlib
plt.imshow(resized_image)
plt.title('Loaded Image')
plt.axis('off')  # Turn off axis labels
plt.show()

# Get image properties
height, width, channels = resized_image.shape  # Shape gives (height, width, channels)
dtype = resized_image.dtype  # Data type of the image (e.g., uint8)
total_pixels = resized_image.size  # Total number of pixels (height * width * channels)

# Print image properties
print(f"Image Dimensions: {width} x {height} pixels")
print(f"Number of Channels: {channels}")  # For grayscale, this will not exist
print(f"Data Type: {dtype}")  # E.g., uint8 (8-bit)


# Save or display the resized image
cv2.imwrite('resized_image.jpg', resized_image)  # Save the resized image
```

```python
# cv2.resize(src, dsize, fx, fy, interpolation)

# src: The source image.
# dsize: Desired output size (width, height) in pixels. If you specify this, fx and fy (scaling factors) will be ignored.
# fx: Scaling factor along the x-axis (width). If dsize is not provided, you can use fx to define the scaling.
# fy: Scaling factor along the y-axis (height). If dsize is not provided, you can use fy to define the scaling.
# interpolation: The interpolation method used for resizing. Common methods include:
    # cv2.INTER_LINEAR (default, good for enlarging)
    # cv2.INTER_NEAREST (fastest, low quality)
    # cv2.INTER_AREA (good for shrinking)
    # cv2.INTER_CUBIC (higher quality for enlarging, slower)
    # cv2.INTER_LANCZOS4 (very high-quality for enlarging, slower)
```

```python
import cv2

# Load image as color
image_color = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)

# OpenCV loads images in BGR format by default, we need to convert it to RGB for correct display in matplotlib
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Resize the image by half (50%) in both directions
resized_image = cv2.resize(image_rgb, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)

# Get image properties
height, width, channels = resized_image.shape  # Shape gives (height, width, channels)
dtype = resized_image.dtype  # Data type of the image (e.g., uint8)
total_pixels = resized_image.size  # Total number of pixels (height * width * channels)

# Print image properties
print(f"Image Dimensions: {width} x {height} pixels")
print(f"Number of Channels: {channels}")  # For grayscale, this will not exist
print(f"Data Type: {dtype}")  # E.g., uint8 (8-bit)

# Plot the image using Matplotlib
plt.imshow(resized_image)
plt.title('Loaded Image')
plt.axis('off')  # Turn off axis labels
plt.show()

```

## Cropping

```python
# Example 1: Cropping a Rectangular Region with Matplotlib
    
import cv2
import matplotlib.pyplot as plt

# 1. Load the image using OpenCV (you can also use plt.imread())
image = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)

# Convert the image from BGR to RGB (OpenCV loads images in BGR, but Matplotlib uses RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 2. Define the region of interest (ROI)
cropped_image = image_rgb[50:300, 100:400]  # Crop from (x1=100, y1=50) to (x2=400, y2=300)

# Get image properties
height, width, channels = cropped_image.shape  # Shape gives (height, width, channels)
dtype = cropped_image.dtype  # Data type of the image (e.g., uint8)
total_pixels = cropped_image.size  # Total number of pixels (height * width * channels)

# Print image properties
print(f"Image Dimensions: {width} x {height} pixels")
print(f"Number of Channels: {channels}")  # For grayscale, this will not exist
print(f"Data Type: {dtype}")  # E.g., uint8 (8-bit)

# 3. Display the cropped image using Matplotlib
plt.imshow(cropped_image)
plt.axis('off')  # Turn off axis labels for a cleaner display
plt.show()
```

```python
# Example 2: Dynamic Cropping and Display with Matplotlib
    
import cv2
import matplotlib.pyplot as plt

# 1. Load the image
image = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# 2. Get the dimensions of the image
height, width = image_rgb.shape[:2]

# 3. Define the region of interest (ROI) dynamically (e.g., crop the center of the image)
start_x = width // 4
start_y = height // 4
end_x = start_x + width // 4
end_y = start_y + height // 4

cropped_image = image_rgb[start_y:end_y, start_x:end_x]

# Get image properties
height, width, channels = cropped_image.shape  # Shape gives (height, width, channels)
dtype = cropped_image.dtype  # Data type of the image (e.g., uint8)
total_pixels = cropped_image.size  # Total number of pixels (height * width * channels)

# Print image properties
print(f"Image Dimensions: {width} x {height} pixels")
print(f"Number of Channels: {channels}")  # For grayscale, this will not exist
print(f"Data Type: {dtype}")  # E.g., uint8 (8-bit)

# 4. Display the cropped image using Matplotlib
plt.imshow(cropped_image)
plt.axis('off')  # Turn off axis labels
plt.show()
```

## Flipping

```python
import cv2
import matplotlib.pyplot as plt

# 1. Load the image using Matplotlib's imread
image_rgb = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)
image = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib


# 2. Flip the image
flipped_vertical = cv2.flip(image, 0)  # Flip vertically
flipped_horizontal = cv2.flip(image, 1)  # Flip horizontally
flipped_both = cv2.flip(image, -1)  # Flip both vertically and horizontally

# 3. Display the original and flipped images using Matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

# Flipped vertically
plt.subplot(1, 4, 2)
plt.imshow(flipped_vertical)
plt.title('Flipped Vertical')
plt.axis('off')

# Flipped horizontally
plt.subplot(1, 4, 3)
plt.imshow(flipped_horizontal)
plt.title('Flipped Horizontal')
plt.axis('off')

# Flipped both
plt.subplot(1, 4, 4)
plt.imshow(flipped_both)
plt.title('Flipped Both')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Rotating

```python
import cv2
import matplotlib.pyplot as plt
from scipy.ndimage import rotate

# 1. Load the image using Matplotlib's imread
image_rgb = cv2.imread('./images/obamas.jpg', cv2.IMREAD_COLOR)
image = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# 2. Rotate the image
angle = 45  # degrees
rotated_image = rotate(image, angle, reshape=True)  # reshape=True allows to adjust the output shape

# 3. Display the original and rotated images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

# Rotated image
plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
